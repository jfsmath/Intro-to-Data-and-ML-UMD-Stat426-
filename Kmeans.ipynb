K-means clustering is a popular unsupervised machine learning algorithm used for partitioning a dataset into K distinct, non-overlapping subsets (clusters). It's widely used in various fields, including data analysis, image segmentation, and customer segmentation. Here's a basic example of how to perform k-means clustering in Python using the scikit-learn library:

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs


n_samples = 200

# Generate sample data, make_blobs is used to generate random data with four clusters
data, _ = make_blobs(n_samples, centers=4, random_state=42)

# Use a different set of data
# np.random.seed(42) #Set a random seed for reproducibility
# data = np.random.rand(n_samples,2)

# Use random samples from a parabola and an outlier
# x = np.random.uniform(0, 1.0, 100)
# y = x**2
# data = np.vstack((x,y)).transpose()
# data = np.vstack((data, np.array([0.2, 0.8])))


# Create KMeans instance with the desired number of clusters (K)
kmeans = KMeans(n_clusters=4)

# Fit the model to the data
kmeans.fit(data)

# Get the cluster centers and labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Plot the original data and cluster centers
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', s = 10, marker = '^')
plt.scatter(centers[:, 0], centers[:, 1], marker='o', s=100, color='red')
plt.title('K-Means Clustering')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans

# Load data into a dataframe
file_path = '/home/black/Documents/Teaching/UMD Stat426 Intro Machine Learning/Code/USJudgeRatings.csv'
df = pd.read_csv(file_path)

# Assuming your DataFrame is named 'df'
# Specify the number of clusters
num_clusters = 4

# Extract the features you want to use for clustering
X = df.select_dtypes(include=['number']).values # Use .values to get the underlying NumPy array

# Initialize the KMeans model
kmeans = KMeans(n_clusters=num_clusters, random_state=42)

# Fit the model to your data
kmeans.fit(X)

# Add a new column to the DataFrame with the cluster labels
df['cluster'] = kmeans.labels_

# Display the DataFrame with cluster labels
#print("DataFrame with Cluster Labels:")
#print(df)
```

```python
import matplotlib.pyplot as plt

feature1 = 'CONT'
feature2 = 'CFMG'

plt.scatter(df[feature1], df[feature2], c=df['cluster'], cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=200, label='Cluster Centers')
plt.title('K-Means Clustering')
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.legend()
plt.show()
```
